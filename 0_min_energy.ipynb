```python
# from sympy import *
from sympy import symbols, solve, diff
```

```python
def energy_min(h, alpha, k, j, m):
    """
    compute the min energy by the given h
    """
    theta1, theta2 = symbols('theta1 theta2')
#     E = k*sin(theta1)**2+k*sin(theta2)**2+j*cos(theta2-theta1)-h*(m*cos(alpha-theta1)+m*cos(alpha-theta2))
    E = 2*theta1**2 - theta2**2+theta2 - 1
    diff_theta1 = diff(E, theta1, 1)
    diff_theta2 = diff(E, theta2, 1)
    
    sol = solve([diff_theta1, diff_theta2], theta1, theta2)
    print(sol)
    print(sol[theta1])
    print(sol[theta2])
    
    return sol, diff_theta1, diff_theta2
```

```python
alpha, k, j, m = 0, 1,2,3
energy_min(3, alpha, k, j, m)
```
